```python
#I divide by type of vehicule- can be use to determine poor, middle, rich class
data['Item Code'] = data['Item Code'].replace(['bicycle'],'bicycle__365days')
data['Item Code'] = data['Item Code'].replace(['motorcycle, scooter'],'motorcycle__365days')
data['Item Code'] = data['Item Code'].replace(['motor car'],'car__365days')

data['Item Code'] = data['Item Code'].replace(['kitchen equipment: sub-total (730-737)'],'kitchen equipment__365days')
data['Item Code'] = data['Item Code'].replace(['crockery and utensils: sub-total (750-753)'],'kitchen utensils__365days')
data['Item Code'] = data['Item Code'].replace(['furniture and fixtures: sub-total (760-768)'],'home furnishing__365days')
data['Item Code'] = data['Item Code'].replace(['IT and communication devices: sub-total (780-787)'],'IT and communication__365days')
data['Item Code'] = data['Item Code'].replace(['miscellaneous durables: sub-total (820-825)'],'Personal_objects__365days')
data['Item Code'] = data['Item Code'].replace(['other machines for household work: sub-total (770-775)'],'household work equipment__365days')
data['Item Code'] = data['Item Code'].replace(['electrical and lighting accessories: sub-total (790-792)'],'electrical and lighting__365days')
data['Item Code'] = data['Item Code'].replace(['jewellery and ornaments: sub-total (830-833)'],'jewellery and ornaments__365days')
data['Item Code'] = data['Item Code'].replace(['sports and medical equipment: sub-total (810-815)'],'sports and medical equipment__365days')
data['Item Code'] = data['Item Code'].replace(['productive equipment: sub-total (800-803)'],'productive equipment__365days')
data['Item Code'] = data['Item Code'].replace(['equipment for recreation: sub-total (740-747)'],'Audio-video__365days')
```
